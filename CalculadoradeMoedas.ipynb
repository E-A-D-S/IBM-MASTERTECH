```python
import requests

print('Olá!\nBem-vindo ao conversor de moedas.\nUtilize o código de cada moeda para fazer a conversão.\nAs moedas suportadas são: \nUSD (dólar);\nBRL (Real); \nGBP (Libra); \nARS (Peso argentino) \nEUR (Euro);\nCAD (Dólar Canadense);\nAUD (Dólar Australiano); \nSGD (Dólar Singapurense);\nCHF (Franco Suíço); \nJPY (Iene Japonês); \nINR (Rúpia Indiana); \nCNY (Yuan Renminb Chinês)')

def get_exchange_rates():
    try:
        url = "https://api.exchangerate-api.com/v4/latest/USD"
        response = requests.get(url)
        response.raise_for_status()  # Check for HTTP errors
        data = response.json()
        return data['rates']
    except requests.exceptions.RequestException as e:
        print(f"Erro ao obter as taxas de câmbio: {e}")
        return None

def converter_moedas(value, currencyOrigin, currencyFinal, rates):
    if currencyFinal in rates and currencyOrigin in rates:
        print('Processando...')
        origin_rate = rates[currencyOrigin]
        destination_rate = rates[currencyFinal]
        usd_value = value / origin_rate
        converted_value = usd_value * destination_rate
        return converted_value
    else:
        print('Moedas não encontradas, tente novamente.')
        return None

exchange_rates = get_exchange_rates()

if exchange_rates is not None:
    while True:
        try:
            value = float(input('\nInforme o valor que deseja converter (ou digite 0 para sair): '))
            if value == 0:
                break

            currencyOrigin = input('\nInforme a Moeda de Origem: ').upper()
            currencyFinal = input('\nInforme a Moeda de Destino: ').upper()

            converted_value = converter_moedas(value, currencyOrigin, currencyFinal, exchange_rates)

            if converted_value is not None:
                print(f'\n{value:.2f} {currencyOrigin} é equivalente a {converted_value:.2f} {currencyFinal}')
        except ValueError:
            print("Valor inválido. Certifique-se de inserir um número válido.")
else:
    print("Não foi possível obter as taxas de câmbio. Verifique sua conexão com a Internet e tente novamente.")





```
